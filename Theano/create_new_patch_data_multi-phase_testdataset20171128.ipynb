```python
path = "/home/dl-box/Liver/sotsuron_train/META"
path1 = "/data1/20171127/Test_dataset/Meta"
image_array11, meta_header11 = mhd_utils_3d.load_raw_data_with_mhd(path+'/ART/ProcessImage/Meta501_ART.mhd')
image_array12, meta_header12 = mhd_utils_3d.load_raw_data_with_mhd(path+'/ART/LiverMask/LiverMask_Srr501_ART.mhd') 
image_array13, meta_header13 = mhd_utils_3d.load_raw_data_with_mhd(path+'/ART/TumorMask/TumorMask_Srr501_ART.mhd')

image_array21, meta_header21 = mhd_utils_3d.load_raw_data_with_mhd(path+'/NC/ProcessImage/Meta501_NC_2.mhd')
image_array31, meta_header31 = mhd_utils_3d.load_raw_data_with_mhd(path+'/PV/ProcessImage/Meta501_PV_2.mhd')
print("Meta501")


import csv
listData=[]
pkl_data ="/data1/20171127/Test_dataset/Meta/Meta501.pkl"

f = open(path1+'/Meta501_location.csv','ab')
Write = csv.writer(f)

data_indx=0
test_listData_intensity=[]
test_listData_label=[]

#new_intensity = np.empty((0,3),int)

###パッチ作成
for z in range(0,image_array11.shape[0]):
    for y in range(0,image_array11.shape[1]):
        for x in range(0,image_array11.shape[2]):       
            if image_array12[z][y][x] == 1:
                data_indx += 1
                new_img1 =  (image_array11[z,y-14:y+15,x-14:x+15])
                if len(new_img1)==0:
                    break
                else:
                    intensity1 = new_img1.reshape((1,841))
                    new_img2 =  (image_array21[z,y-14:y+15,x-14:x+15])
                    intensity2 = new_img2.reshape((1,841))
                    new_img3 =  (image_array31[z,y-14:y+15,x-14:x+15])
                    intensity3 = new_img3.reshape((1,841))
                    
                    new_img4 = [intensity1[0],intensity2[0],intensity3[0]]
                    new_img4 = np.array([new_img4]).transpose(0,2,1)
                    new_img5 = new_img4 
                    label_L = 0
                    
                    test_listData_intensity.append(new_img5)
                    test_listData_label.append(label_L)
                    listData.append(np.array([x,y,z]))
                    
for i in range(len(listData)):
    Write.writerow((listData[i]))
f.close()    
                    
print("dump!")
    
test_listData=[]
listData=[]

test_listData.append(test_listData_intensity)
test_listData.append(test_listData_label)

listData.append(test_listData)

pickle.dump(listData, open(pkl_data,"w"))
```
